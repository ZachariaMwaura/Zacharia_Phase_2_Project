##### Imports


```python

import pandas as pd
```

--------

##### Read in the files

```python
# Directors

directors = pd.read_csv('DATA/Directors.csv', index_col=False)
directors.sample(n=10, random_state=1)
```

```python
# tmdb file

TMDB1 = pd.read_csv('DATA/tmdb_movies_1.csv')
TMDB1.sample(n=10, random_state=1)
```

```python
TMDB2 = pd.read_csv('DATA/tmdb_movies_2.csv')
TMDB2.sample(n=10, random_state=1)
```

```python
TMDB3 = pd.read_csv('DATA/tmdb_movies_3.csv')
TMDB3.sample(n=10, random_state=1)
```

```python
# concatinate the three TMDB files

TMDB_FULL = pd.concat([TMDB1, TMDB2, TMDB3], ignore_index=True)
```

```python
TMDB_FULL.info()
```

```python
TMDB_FULL.to_csv('CLEANDATA/tmdb_full.csv', index=False)
```
